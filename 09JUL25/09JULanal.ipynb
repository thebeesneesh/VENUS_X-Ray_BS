```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('errortest.csv')

plt.scatter(df['Total Power'], df['Spectral Temperature'])
plt.xlabel('Total Power (W)')
plt.ylabel('Spectral Temperature (eV)')
plt.title('Power vs Ts')
plt.show()
```

```python
plt.scatter(df['Puller I'], df['Spectral Temperature'])
plt.xlabel('Puller Drain Current (A)')
plt.ylabel('Spectral Temperature (eV)')
plt.title('Puller Drain Current vs Ts')
plt.show()
```

```python
plt.scatter(df['Inj Pres (-7)'], df['Spectral Temperature'])
plt.xlabel('Injection Pressure (uTorr)')
plt.ylabel('Spectral Temperature (eV)')
plt.title('Pressure vs Ts')
plt.show()
```

```python
plt.scatter(df['Bias I'], df['Spectral Temperature'])
plt.xlabel('Bias Current (A)')
plt.ylabel('Spectral Temperature (eV)')
plt.title('Bias Current vs Ts')
plt.show()
```

```python
plt.scatter(df['18/14'], df['Spectral Temperature'], label ='18/14')
plt.scatter(df['28/18'], df['Spectral Temperature'], label ='28/18')
plt.scatter(df['28/14'], df['Spectral Temperature'], label ='28/14')
plt.xlabel('Power Ratio')
plt.ylabel('Spectral Temperature (eV)')
plt.legend()
plt.title('Power Ratios vs Ts')
plt.show()
```

```python
#plt.scatter(df['Total Counts'], df['Spectral Temperature'], label = 'Total')
#plt.scatter(df['Norm Counts'], df['Spectral Temperature'], label = 'Norm')
fix, ax1 = plt.subplots()
color = 'tab:red'
ax1.set_xlabel('Total Counts')
ax1.set_ylabel('Spectral Temperature (eV)')
ax1.scatter(df['Total Counts'], df['Spectral Temperature'], label = 'Total', color = 'red')
ax2 = ax1.twiny()
color = 'tab:blue'
ax2.scatter(df['Norm Counts'], df['Spectral Temperature'], label = 'Norm', color = 'blue')
ax2.set_xlabel('Counts / sec')
plt.title('Total Counts vs Ts')
plt.legend()
plt.show()
```

```python
plt.scatter(df['Norm Counts'], df['Spectral Temperature'])
plt.xlabel('Counts / sec')
plt.ylabel('Spectral Temperature (eV)')
plt.title('Normalized Counts vs Ts')
plt.show()
```

```python
plt.scatter(df['Total Power'], df['Total Counts'], label = 'Total Counts')
plt.xlabel('Total Power (W)')
plt.ylabel('Total Counts')
plt.title('Total Counts vs Total Power')
plt.legend()
plt.show()
```

```python
plt.scatter(df['Total Power'], df['Norm Counts'], label = 'Norm Counts')
plt.xlabel('Total Power (W)')
plt.ylabel('Counts / sec')
plt.title('Norm Counts vs Total Power')
plt.legend()
plt.show()
```

```python
plt.scatter(df['Total Power'], df['Total Counts'])
plt.xlabel('Power (W)')
plt.ylabel('Counts')
plt.title('Counts vs Power')
plt.show()
```

```python
plt.errorbar(df['Total Counts'], df['New Ts'], yerr = df['Error'], xerr=np.sqrt(df['Total Counts']), fmt = 'k*', capsize = 2)
plt.xlabel('Counts')
plt.ylabel('Ts')
plt.title('Error')
plt.legend()
plt.show()
```
