```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('10juldatanew.csv')

plt.scatter(df['Total Power Ideal'], df['Ts'])
plt.xlabel('Total Power (W)')
plt.ylabel('Spectral Temperature (eV)')
plt.title('Power vs Ts')
plt.show()
```

```python
plt.scatter(df['Bmin (Panelview)'], df['Ts'])
plt.xlabel('Bmin (T)')
plt.ylabel('Spectral Temperature (eV)')
plt.title('Bmin vs Ts')
plt.show()
```

```python
plt.scatter(df['Input Counts'], df['Ts'])
plt.xlabel('Counts')
plt.ylabel('Spectral Temperature (eV)')
#plt.scatter(df['Bmin (Panelview)'], color = 'k', df['Ts'], color = 'red')
plt.title('Counts vs Ts')
plt.show()
```

```python
all3 = df[(df['14 GHz Ideal'] == 500) & (df['18GHz Ideal'] == 500) & (df['28GHz Ideal'] == 500)]
sub_1828G = df[(df['14 GHz Ideal'] == 0) & (df['18GHz Ideal'] == 500) & (df['28GHz Ideal'] == 500)]
sub_1428G = df[(df['14 GHz Ideal'] == 500) & (df['18GHz Ideal'] == 0) & (df['28GHz Ideal'] == 500)]
sub_28G = df[(df['14 GHz Ideal'] == 0) & (df['18GHz Ideal'] == 0) & (df['28GHz Ideal'] == 500)]

plt.plot(sub_28G['Bmin (Panelview)'], sub_28G['Ts'], marker = 'o', label = '28 GHz only (500 W)')
plt.plot(sub_1828G['Bmin (Panelview)'], sub_1828G['Ts'], marker = 'o', label = '18 + 28 GHz (1000 W)')
plt.plot(sub_1428G['Bmin (Panelview)'], sub_1428G['Ts'], marker = 'o', label = '14 + 28 GHz (1000 W)')
plt.plot(all3['Bmin (Panelview)'], all3['Ts'], marker = 'o', label = '14 + 18 + 28 GHz (1500 W)')

plt.xlabel('Bmin (T)')
plt.ylabel('Ts')
plt.title('Bmin vs Ts')
plt.legend()
plt.show()
```

Bmin value recorded from Panelview
much less heating with 14 + 28 than with 18 + 28
only marginally more heating with 18 + 28 than just 28
weirdly all kind of the same at very high Bmin (good count rate)

```python
plt.plot(sub_1828G['Bmin (Panelview)'], sub_1828G['Ts'], marker = 'o', label = '18 + 28 GHz (1000 W)')
plt.plot(sub_1428G['Bmin (Panelview)'], sub_1428G['Ts'], marker = 'o', label = '14 + 28 GHz (1000 W)')

plt.xlabel('Bmin (T)')
plt.ylabel('Ts')
plt.title('Bmin vs Ts')
plt.legend()
plt.show()
```

would expect Ts to rise linearly with Bmin, but not the case for 14 + 28 (or anything with 14)

```python
plt.plot(sub_28G['Bmin (Panelview)'], sub_28G['Norm Ts'], marker = 'o', label = '28 GHz only (500 W)')
plt.plot(sub_1828G['Bmin (Panelview)'], sub_1828G['Norm Ts'], marker = 'o', label = '18 + 28 GHz (1000 W)')
plt.plot(sub_1428G['Bmin (Panelview)'], sub_1428G['Norm Ts'], marker = 'o', label = '14 + 28 GHz (1000 W)')
plt.plot(all3['Bmin (Panelview)'], all3['Norm Ts'], marker = 'o', label = '14 + 18 + 28 GHz (1500 W)')

plt.xlabel('Bmin (T)')
plt.ylabel('Ts / W')
plt.title('Bmin vs Ts / W')
plt.legend()
plt.show()
```
