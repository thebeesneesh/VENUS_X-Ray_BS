```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('20may.csv')

plt.scatter(df['28 GHz'], df['Spectral Temp'])
plt.xlabel('28 GHz Power (W)')
plt.ylabel('Spectral Temperature (eV)')
plt.title('28 GHz vs Ts')
plt.show()
```

```python
plt.scatter(df['Inj Pres (-7)'], df['Spectral Temp'])
plt.xlabel('Pressure')
plt.ylabel('Spectral Temperature (eV)')
plt.title('Inj Pressure vs Ts')
plt.show()
```

```python
plt.scatter(df['Cent (T)'], df['Spectral Temp'])
plt.xlabel('Bmin')
plt.ylabel('Spectral Temperature (eV)')
plt.title('Bmin vs Ts')
plt.show()
```

```python
plt.scatter(df['Sext (T)'], df['Spectral Temp'])
plt.xlabel('Brad')
plt.ylabel('Spectral Temperature (eV)')
plt.title('Brad vs Ts')
plt.show()
```

```python
plt.errorbar(df['O6+ I (uA)'], df['Spectral Temp'], yerr = df['Error'], fmt = 'k*', capsize = 2)
plt.xlabel('Current (uA)')
plt.ylabel('Ts')
plt.title('Error')
#plt.legend()
plt.show()
```
