```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('10juldata.csv')

plt.scatter(df['Total Power Ideal'], df['Ts'])
plt.xlabel('Total Power (W)')
plt.ylabel('Spectral Temperature (eV)')
plt.title('Power vs Ts')
plt.show()
```

```python
plt.scatter(df['Bmin (Panelview)'], df['Ts'])
plt.xlabel('Bmin (T)')
plt.ylabel('Spectral Temperature (eV)')
plt.title('Bmin vs Ts')
plt.show()
```

```python
plt.scatter(df['Input Counts'], df['Ts'])
plt.xlabel('Counts')
plt.ylabel('Spectral Temperature (eV)')
#plt.scatter(df['Bmin (Panelview)'], color = 'k', df['Ts'], color = 'red')
plt.title('Counts vs Ts')
plt.show()
```

```python
plt.scatter(df['18/14 Ratio Ideal'], df['Ts'], label = '18/14')
plt.scatter(df['28/14 Ratio Ideal'], df['Ts'], label = '28/14')
plt.scatter(df['28/18 Ratio Ideal'], df['Ts'], label = '28/18')
plt.xlabel('Ratio')
plt.ylabel('Spectral Temperature (eV)')
plt.title('Power Ratios vs Ts')
plt.legend()
plt.show()
```

```python
plt.scatter(df['18GHz Ideal'], df['Ts'], label = '18 GHz power')
plt.scatter(df['14 GHz Ideal'], df['Ts'], label = '14 GHz power')
plt.scatter(df['28GHz Ideal'], df['Ts'], label = '28 GHz power')
plt.xlabel('Power')
plt.ylabel('Spectral Temperature (eV)')
plt.title('Power Ratios vs Ts')
plt.legend()
plt.show()
```

```python
all3 = df[(df['14 GHz Ideal'] == 500) & (df['18GHz Ideal'] == 500) & (df['28GHz Ideal'] == 500)]
sub_1828G = df[(df['14 GHz Ideal'] == 0) & (df['18GHz Ideal'] == 500) & (df['28GHz Ideal'] == 500)]
sub_1428G = df[(df['14 GHz Ideal'] == 500) & (df['18GHz Ideal'] == 0) & (df['28GHz Ideal'] == 500)]
sub_28G = df[(df['14 GHz Ideal'] == 0) & (df['18GHz Ideal'] == 0) & (df['28GHz Ideal'] == 500)]


plt.plot(sub_28G['Bmin (Panelview)'], sub_28G['Ts'], marker = 'o', label = '28 GHz only')
plt.plot(sub_1828G['Bmin (Panelview)'], sub_1828G['Ts'], marker = 'o', label = '18 + 28 GHz')
plt.plot(sub_1428G['Bmin (Panelview)'], sub_1428G['Ts'], marker = 'o', label = '14 + 28 GHz')
plt.plot(all3['Bmin (Panelview)'], all3['Ts'], marker = 'o', label = '14 + 18 + 28 GHz')

plt.xlabel('Bmin (T)')
plt.ylabel('Ts')
plt.title('Bmin vs Ts')
plt.legend()
plt.show()
```
